### What is $1 collected T years from now (future value) worth today (present value)? 

The value of money is not constant over time. Thus, one dollar in Year T is worth less than one dollar in Year 0 (today). The decay in the value of money is called the time value of money. 

A useful mental model is to think about money from different time periods as being in different currencies (e.g., dollars and euros). Thus, to compare money from different time periods we formulate the equivalent of exchange rates. These exchange rates (from time $j$ to $j+1$) are called discount factors $r_{j+1,j}$. Using these factors we can formulate an exchange between cash flow at time 1 (now) and some case flow at time $i$ in the future:

$$CF_{i} = CF_{1}\cdot\left[\prod_{j=1}^{i-1}(1+r_{j+1,j})\right]\qquad{i=2,3,\dots{T}}$$

where the term in brackets is the conversion (or discount) from now an some point in the future. Using this idea, let's look at how much a dollar in the future is worth today.

```julia
# include required packages -
using PyPlot
```

```julia
# setup some constants -

# simplification: assume the discount rate is constant between now and T -
discount_rate_array = [0.04,0.08,0.12];
number_of_discount_rates = length(discount_rate_array)

# how many time periods (e.g, years) in the future do we want to look?
number_of_time_periods = 30;
```

```julia
# initialize -
present_value_array = ones(number_of_time_periods,number_of_discount_rates)

# main loop -
for discount_rate_index = 1:number_of_discount_rates 
    for time_period_index = 2:number_of_time_periods
    
        # init the discount term -
        discount_term = 1.0
    
        # compute the discount term -
        for discount_index = 1:(time_period_index - 1)
            discount_term = discount_term*(1+discount_rate_array[discount_rate_index])
        end
    
        # compute the present value -
        tmp_term = 1.0/(discount_term)

        # store -
        present_value_array[time_period_index,discount_rate_index] = tmp_term
    end
end
```

```julia
# make a plot -
plot(present_value_array[:,1],label="discount rate = 0.04");
plot(present_value_array[:,2],label="discount rate = 0.08");
plot(present_value_array[:,3],label="discount rate = 0.12");

# add labels -
xlabel("Time perod index",fontsize=16);
ylabel("Present value of \$1 T years from now",fontsize=14);

# Add a legend -
legend(loc="upper right");
```
